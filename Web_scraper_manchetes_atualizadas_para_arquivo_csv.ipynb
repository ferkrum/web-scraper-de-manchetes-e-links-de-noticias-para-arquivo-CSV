# Web scraper de manchetes atualizadas da Folha de São Paulo para arquivo CSV

Como obter uma lista com os títulos e links das notícias publicadas na página principal da Folha de São Paulo? Como exportar o resultado para um arquivo separado por vírgulas (CSV)?



Linguagem: Python

_libs: requests; Beautifulsoup; pandas_


Referêcias:
* Beautifulsoup [aqui](https://beautiful-soup-4.readthedocs.io/en/latest/).

```python
import requests
from bs4 import BeautifulSoup
```

```python
from datetime import datetime
from pytz import timezone 
brasil = timezone('Brazil/East')
bs_time = datetime.now(brasil)
timestampBrasil = bs_time.strftime('%Y-%m-%d_%H-%M-%S')
#print(bs_time)
```

```python
response = requests.get('https://www.folha.uol.com.br/').content
```

```python
#parser
soup = BeautifulSoup(response, 'html.parser')
```

```python
soup.find_all("h2",class_= "c-main-headline__title")
```

```python
noticias = soup.find_all("a",class_= "c-headline__url")
```

```python
noticias[0]
```

```python
dictNoticias = {}
contador = 0 
for i in noticias:
  contador+=1
  #print(contador)
  try: 
    tituloNoticia = i.find("h2").get_text(' ', strip = True) #funcao para remover espaços da string
  except:
    tituloNoticia = 'none'
  urlNoticia = i['href']
  try:
    mancheteNoticia = i.find("p").get_text(' ', strip = True)
  except:
    mancheteNoticia = 'none'
  #cria dicionario
  dictNoticias[contador] = {"titulo" : tituloNoticia, "manchete" : mancheteNoticia, "url" : urlNoticia}
```

```python
import pandas as pd
```

```python
dfNoticias = pd.DataFrame.from_dict(dictNoticias).T
pd.DataFrame.from_dict(dictNoticias).T.to_csv('noticias_folha' + timestampBrasil + '.csv')
```
